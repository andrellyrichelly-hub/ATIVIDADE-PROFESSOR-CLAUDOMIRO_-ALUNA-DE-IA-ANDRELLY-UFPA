Questão 7 - Slide 88

Resposta:

1. Barramento de Endereços (BE)

- Função: Transporta o endereço da memória que será acessada. Conecta o REM à MP.
- Direção: Unidirecional - apenas da CPU para a MP.
2. Barramento de Dados (BD)

- Função: Transporta dados e instruções entre CPU e MP. Conecta o RDM à MP.
- Direção: Bidirecional - CPU ↔ MP.

3. Barramento de Controle (BC)

- Função: Transporta sinais de controle como Read, Write, Clock e Interrupção.
- Direção: Bidirecional - CPU ↔ MP.

```python
# Questão 13 - Slide 88
endereco_inicial_hex = "27FA"
endereco_inicial_dec = int(endereco_inicial_hex, 16)
tamanho_grupo = 128
posicao_J = 9  # A=0, B=1, ..., J=9

endereco_1J_dec = endereco_inicial_dec + (posicao_J * tamanho_grupo)
endereco_1J_hex = hex(endereco_1J_dec).upper().replace('0X', '')

print(f"Resposta Questão 13: O 1º J está no endereço {endereco_1J_hex}")
# Saída: 2C7A
```

```python
# Questão 14 - Slide 88
interface_DRAM = 5.00
interface_SRAM = 1.00
preco_bit_SRAM = 0.00002
preco_bit_DRAM = 0.00001

N_minimo = (interface_DRAM - interface_SRAM) / (preco_bit_SRAM - preco_bit_DRAM)

print(f"Resposta Questão 14: DRAM compensa acima de {N_minimo:,.0f} bits")
print(f"Ou seja: {N_minimo/8:,.0f} bytes = {N_minimo/(8*1024):.1f} Kbytes")
# Saída: 400,000 bits = 50,000 bytes = 48.8 Kbytes
```
