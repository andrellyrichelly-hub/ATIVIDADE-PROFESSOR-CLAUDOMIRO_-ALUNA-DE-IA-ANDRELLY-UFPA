```python
# Exemplo 4.3 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
N_palavras = 2 * 1024  # 2K palavras = 2 * 1024 = 2048 palavras/células
M_bits_por_palavra = 16  # cada palavra/célula tem 16 bits

# 1. Tamanho do REM - Registrador de Endereços da Memória
# REM guarda o endereço. Precisa ter bits suficientes pra endereçar N posições
import math
REM_bits = int(math.log2(N_palavras))

# 2. Tamanho do RDM - Registrador de Dados da Memória
# RDM guarda o dado que vai/volta da memória. Tem que ter o mesmo tamanho da célula
RDM_bits = M_bits_por_palavra

# 3. Valor do maior endereço
# Endereços vão de 0 até N-1
maior_endereco_decimal = N_palavras - 1
maior_endereco_binario = bin(maior_endereco_decimal)
maior_endereco_hex = hex(maior_endereco_decimal)

# 4. Quantidade total de bits armazenados
T_total_bits = N_palavras * M_bits_por_palavra
T_total_Kbits = T_total_bits / 1024
T_total_bytes = T_total_bits / 8

print("=== Resultado Exemplo 4.3 ===")
print(f"1. Tamanho do REM: {REM_bits} bits")
print(f"   → Porque precisa endereçar {N_palavras} posições: 2^{REM_bits} = {2**REM_bits}")
print(f"\n2. Tamanho do RDM: {RDM_bits} bits")
print(f"   → Porque cada célula/palavra da MP tem {M_bits_por_palavra} bits")
print(f"\n3. Maior endereço da MP: {maior_endereco_decimal} em decimal")
print(f"   → Em binário: {maior_endereco_binario}")
print(f"   → Em hexa: {maior_endereco_hex}")
print(f"\n4. Total de bits na MP: {T_total_bits} bits")
print(f"   = {T_total_Kbits} Kbits = {T_total_bytes} bytes = 4 Kbytes")
```
