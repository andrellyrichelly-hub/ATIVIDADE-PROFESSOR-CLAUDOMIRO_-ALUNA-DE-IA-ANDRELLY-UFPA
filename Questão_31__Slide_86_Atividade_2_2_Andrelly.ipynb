```python
# Questão 31 - Slide 86 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
largura = 640  # pontos
altura = 420   # pontos
bytes_por_pixel = 4  # 4 bytes por ponto/cor = 32 bits = True Color
qtd_imagens = 10
RAM_total_MB = 128

# a) Quantidade de memória para 1 imagem em bytes
total_pixels = largura * altura
memoria_1_imagem_bytes = total_pixels * bytes_por_pixel
memoria_1_imagem_KB = memoria_1_imagem_bytes / 1024
memoria_1_imagem_MB = memoria_1_imagem_KB / 1024

# b) Memória para 10 imagens em megabytes
memoria_10_imagens_bytes = memoria_1_imagem_bytes * qtd_imagens
memoria_10_imagens_MB = memoria_10_imagens_bytes / (1024**2)

# c) Quantas imagens cabem em 128MB de RAM
RAM_total_bytes = RAM_total_MB * (1024**2)
qtd_imagens_na_RAM = RAM_total_bytes // memoria_1_imagem_bytes  # divisão inteira

print("=== Resultado Questão 31 ===")
print(f"Resolução: {largura} × {altura} = {total_pixels:,} pixels")
print(f"Profundidade de cor: {bytes_por_pixel} bytes/pixel = {bytes_por_pixel*8} bits\n")

print(f"a) Memória para 1 imagem:")
print(f"   = {total_pixels:,} pixels × {bytes_por_pixel} bytes")
print(f"   = {memoria_1_imagem_bytes:,} bytes")
print(f"   = {memoria_1_imagem_KB:,.2f} Kbytes")
print(f"   = {memoria_1_imagem_MB:.3f} Mbytes\n")

print(f"b) Memória para 10 imagens:")
print(f"   = {memoria_1_imagem_bytes:,} × 10 = {memoria_10_imagens_bytes:,} bytes")
print(f"   = {memoria_10_imagens_MB:.2f} Mbytes\n")

print(f"c) Quantas imagens em 128MB de RAM:")
print(f"   128MB = {RAM_total_bytes:,} bytes")
print(f"   {RAM_total_bytes:,} ÷ {memoria_1_imagem_bytes:,} = {qtd_imagens_na_RAM:.0f} imagens completas")
print(f"   Sobra: {RAM_total_bytes % memoria_1_imagem_bytes:,} bytes sem uso")
```
