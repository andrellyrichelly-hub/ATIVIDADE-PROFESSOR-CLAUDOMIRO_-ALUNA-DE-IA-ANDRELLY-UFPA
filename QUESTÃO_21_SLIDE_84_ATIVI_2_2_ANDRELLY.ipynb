```python
# Questão 21 - Slide 84 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
T_cache = 8  # Tempo processador → cache: 8 ns
T_mp = 70    # Tempo cache → memória principal: 70 ns
n_acessos = 100  # 100 acessos consecutivos
taxa_acerto = 0.96  # Eficiência/acerto = 96% = 0.96
taxa_erro = 1 - taxa_acerto  # Taxa de miss = 4% = 0.04

# Fórmula do Tempo Médio de Acesso - TMA
# TMA = (Taxa_Acerto × T_cache) + (Taxa_Erro × Tempo_Penalidade)
# Quando tem MISS, o tempo é T_cache + T_mp, porque primeiro tenta na cache e depois vai na MP
T_penalidade = T_cache + T_mp

# Calculando o TMA
TMA = (taxa_acerto * T_cache) + (taxa_erro * T_penalidade)

# Tempo total pra 100 acessos
T_total_100_acessos = n_acessos * TMA

# Só pra conferir: calculando na mão quantos acertos e erros
acertos = n_acessos * taxa_acerto
erros = n_acessos * taxa_erro
T_total_manual = (acertos * T_cache) + (erros * T_penalidade)

print("=== Resultado Questão 21 ===")
print(f"Dados: T_cache = {T_cache} ns | T_MP = {T_mp} ns | Taxa acerto = {taxa_acerto*100}%")
print(f"\n1. Tempo de penalidade quando dá MISS:")
print(f"   T_penalidade = T_cache + T_MP = {T_cache} + {T_mp} = {T_penalidade} ns")
print(f"\n2. Tempo Médio de Acesso - TMA:")
print(f"   TMA = (Acerto × T_cache) + (Erro × T_penalidade)")
print(f"   TMA = ({taxa_acerto} × {T_cache}) + ({taxa_erro} × {T_penalidade})")
print(f"   TMA = {taxa_acerto * T_cache} + {taxa_erro * T_penalidade} = {TMA} ns")
print(f"\n3. Para 100 acessos consecutivos:")
print(f"   Tempo total = 100 × {TMA} = {T_total_100_acessos} ns")
print(f"\nConferindo:")
print(f"   Acertos: {acertos:.0f} × {T_cache} ns = {acertos * T_cache} ns")
print(f"   Erros: {erros:.0f} × {T_penalidade} ns = {erros * T_penalidade} ns")
print(f"   Total: {T_total_manual} ns ✓")
```
