```python
# Questão 1 - Slides 77/78 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
BE_bits = 12  # Barramento de Endereços = 12 bits → REM = 12 bits
palavra_bits = 16  # Cada célula armazena 1 palavra de 16 bits
bits_por_byte = 8  # Conversão padrão: 1 byte = 8 bits

# 1. Quantidade de células/endereços
# BE = 12 bits, então N = 2^12
N_total_celulas = 2 ** BE_bits

# 2. Total de bits da memória
T_total_bits = N_total_celulas * palavra_bits

# 3. Convertendo pra bytes
T_total_bytes = T_total_bits // bits_por_byte
T_total_Kbytes = T_total_bytes / 1024

print("=== Resultado Questão 1 ===")
print(f"1. Barramento de Endereços: {BE_bits} bits")
print(f"   → Total de células/endereços: 2^{BE_bits} = {N_total_celulas:,} células")
print(f"\n2. Cada célula armazena: {palavra_bits} bits = {palavra_bits // bits_por_byte} bytes")
print(f"\n3. Total armazenado na memória:")
print(f"   = {N_total_celulas:,} células × {palavra_bits} bits")
print(f"   = {T_total_bits:,} bits")
print(f"   = {T_total_bytes:,} bytes")
print(f"   = {T_total_Kbytes} Kbytes = 8 Kbytes")
```
