```python
# Exemplo 4.1 - Atividade 2.2 - Arquitetura de Computadores
# Aluno: Adrelly Richelly

# Dados do problema
N = 2 * 1024  # 2K = 2 * 1024 células = 2048 células
M = 16        # cada célula armazena 16 bits

# 1. Valor total de bits da memória
T_total_bits = N * M
T_total_Kbits = T_total_bits / 1024
T_total_bytes = T_total_bits / 8
T_total_Kbytes = T_total_bytes / 1024

# 2. Tamanho de cada endereço em bits
# Fórmula: N = 2^E, onde E = quantidade de bits do endereço
import math
E_bits_endereco = int(math.log2(N))

print("=== Resultado Exemplo 4.1 ===")
print(f"1. Total de bits armazenados: {T_total_bits} bits")
print(f"   = {T_total_Kbits} Kbits")
print(f"   = {T_total_bytes} bytes = {T_total_Kbytes} Kbytes")
print(f"\n2. Tamanho de cada endereço: {E_bits_endereco} bits")
print(f"   Porque 2^{E_bits_endereco} = {2**E_bits_endereco} endereços")
```
