```python
# ============================================================================
# ATIVIDADE 2.4 - PROJETOS SOBRE MEMÓRIAS DO TANENBAUM
# ALUNA: Andrelly
# DISCIPLINA: Organização e Arquitetura de Computadores
# PROFESSOR: Claudomiro de Souza de Sales Junior
# DATA: 22/06/2026
# ============================================================================

print("="*80)
print("ATIVIDADE 2.4 - PROJETOS SLIDES 19 E 20 | Andrelly".center(80))
print("="*80)

# =============================================================================
# PROJETO SLIDE 19
# =============================================================================
print("\n" + "="*80)
print("PROJETO SLIDE 19".center(80))
print("="*80)

# Requisitos do slide 19
print("\n>> REQUISITOS:")
print("   1. 13 pinos de endereços")
print("   2. BD transporta 2 palavras, máx 64 bits")
print("   3. Palavra > célula da memória")
print("   4. Capacidade de 8 Gbits")
print("   5. Indicar palavra, célula, BD, barramento de endereços + desenhar chip")

# Cálculos
p19_pinos_end = 13
p19_bd_max = 64
p19_palavras_acesso = 2
p19_capacidade_gbits = 8
p19_celula_bits = 8

p19_total_end = 2 ** p19_pinos_end
p19_capacidade_bits = p19_capacidade_gbits * (2**30)
p19_bits_por_end = p19_capacidade_bits // p19_total_end
p19_palavra_bits = p19_bd_max // p19_palavras_acesso
p19_bd_bits = p19_palavra_bits * p19_palavras_acesso

# Validações
assert p19_pinos_end == 13, "Slide 19: Pinos != 13"
assert p19_bd_bits <= 64, "Slide 19: BD > 64"
assert p19_palavra_bits > p19_celula_bits, "Slide 19: Palavra <= Célula"
assert p19_total_end * p19_bits_por_end == p19_capacidade_bits, "Slide 19: Capacidade errada"

# Resultados
print("\n" + "-"*80)
print(">> RESULTADOS PROJETO 19:")
print(f"   Barramento de Endereços: {p19_pinos_end} bits -> {p19_total_end} endereços")
print(f"   Tamanho da Célula:       {p19_celula_bits} bits = 1 byte")
print(f"   Tamanho da Palavra:      {p19_palavra_bits} bits = {p19_palavra_bits//8} bytes")
print(f"   Barramento de Dados BD:  {p19_bd_bits} bits -> {p19_palavras_acesso} palavras")
print(f"   Bits por Endereço:       {p19_bits_por_end} bits = {p19_bits_por_end//8} bytes")
print(f"   Capacidade Total:        {p19_capacidade_gbits} Gbits - OK")

# Desenho Chip 19
print("\n>> DESENHO DO CHIP - SLIDE 19")
print("                 +------------------+")
print(f"     A0 ---------|                  |---------- D0")
print(f"     A1 ---------|   MEMÓRIA        |---------- D1")
print(f"     ...         |   8K x 1Mbit     |   ...")
print(f"     A12---------|   8 Gbits        |---------- D63")
print("                 |                  |")
print("     CS ---------|                  |")
print("     OE ---------|                  |")
print("     WE ---------|                  |")
print("                 +------------------+")

print("\n>> JUSTIFICATIVA 19:")
print("   Com 13 pinos e 8 Gbits, cada endereço tem 1 Mbit.")
print("   Palavra 32 bits: 2x32=64 respeita BD<=64 e é > célula de 8 bits.")

# =============================================================================
# PROJETO SLIDE 20
# =============================================================================
print("\n\n" + "="*80)
print("PROJETO SLIDE 20".center(80))
print("="*80)

# Requisitos do slide 20
print("\n>> REQUISITOS:")
print("   1. 16 linhas de dados (BD)")
print("   2. Pinos de endereço < 14")
print("   3. Capacidade de 32 Gbits")
print("   4. Justificar decisões + desenhar chip")
print("   5. Qual tamanho do barramento de endereços?")

# Cálculos
p20_bd_bits = 16
p20_max_pinos = 14
p20_capacidade_gbits = 32
p20_celula_bits = 8

p20_pinos_end = 13  # máximo < 14
p20_total_end = 2 ** p20_pinos_end
p20_capacidade_bits = p20_capacidade_gbits * (2**30)
p20_bits_por_end = p20_capacidade_bits // p20_total_end
p20_palavra_bits = p20_bd_bits

# Validações
assert p20_pinos_end < p20_max_pinos, "Slide 20: Pinos >= 14"
assert p20_palavra_bits > p20_celula_bits, "Slide 20: Palavra <= Célula"
assert p20_total_end * p20_bits_por_end == p20_capacidade_bits, "Slide 20: Capacidade errada"

# Resultados
print("\n" + "-"*80)
print(">> RESULTADOS PROJETO 20:")
print(f"   Barramento de Endereços: {p20_pinos_end} bits -> {p20_total_end} endereços")
print(f"   RESPOSTA:                {p20_pinos_end} bits, pois é < 14")
print(f"   Barramento de Dados BD:  {p20_bd_bits} bits")
print(f"   Tamanho da Palavra:      {p20_palavra_bits} bits = {p20_palavra_bits//8} bytes")
print(f"   Tamanho da Célula:       {p20_celula_bits} bits = 1 byte")
print(f"   Bits por Endereço:       {p20_bits_por_end} bits = {p20_bits_por_end//8} bytes")
print(f"   Capacidade Total:        {p20_capacidade_gbits} Gbits - OK")

# Desenho Chip 20
print("\n>> DESENHO DO CHIP - SLIDE 20")
print("                 +------------------+")
print(f"     A0 ---------|                  |---------- D0")
print(f"     A1 ---------|   MEMÓRIA        |---------- D1")
print(f"     ...         |   8K x 4Mbit     |   ...")
print(f"     A12---------|   32 Gbits       |---------- D15")
print("                 |                  |")
print("     CS ---------|                  |")
print("     OE ---------|                  |")
print("     WE ---------|                  |")
print("                 +------------------+")

print("\n>> JUSTIFICATIVA 20:")
print("   Usei 13 pinos por ser máximo < 14. Assim minimizo bits por endereço.")
print("   Com 13 pinos e 32 Gbits, cada endereço tem 4 Mbits.")
print("   Palavra 16 bits = BD e é > célula de 8 bits.")

# =============================================================================
# RESUMO FINAL
# =============================================================================
print("\n\n" + "="*80)
print("RESUMO FINAL - ATIVIDADE 2.4".center(80))
print("="*80)
print("\n| Projeto  | Barr. End | BD      | Palavra | Célula | Capacidade |")
print("|----------|-----------|---------|---------|--------|------------|")
print(f"| Slide 19 | {p19_pinos_end} bits   | {p19_bd_bits} bits | {p19_palavra_bits} bits | {p19_celula_bits} bits | {p19_capacidade_gbits} Gbits    |")
print(f"| Slide 20 | {p20_pinos_end} bits   | {p20_bd_bits} bits | {p20_palavra_bits} bits | {p20_celula_bits} bits | {p20_capacidade_gbits} Gbits   |")
print("\n" + "="*80)
print("STATUS: 2/2 projetos validados conforme Tanenbaum. Pronto pro GitHub.")
print("="*80)
```
