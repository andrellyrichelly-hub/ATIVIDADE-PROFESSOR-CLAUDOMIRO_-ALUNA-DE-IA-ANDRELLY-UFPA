```python
# Questão 10 - Slide 81 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
RDM_bits = 16  # RDM = 16 bits → tamanho da palavra
REM_bits = 20  # REM = 20 bits → capacidade de endereçamento
celula_bits = 8  # Cada célula = 8 bits = 1 byte

# a) Tamanho do barramento de endereços (BE)
# O BE tem sempre o mesmo tamanho do REM, porque eles são conectados
BE_bits = REM_bits

# b) Quantas células são lidas em uma única operação de leitura?
# RDM = 16 bits = palavra. Cada célula = 8 bits.
# Então pra formar 1 palavra, preciso ler: RDM_bits / celula_bits
celulas_por_leitura = RDM_bits // celula_bits

# c) Quantos bits tem a memória principal?
# REM = 20 bits → consigo endereçar 2^20 células
N_total_celulas = 2 ** REM_bits
T_total_bits = N_total_celulas * celula_bits
T_total_bytes = T_total_bits / 8
T_total_Mbytes = T_total_bytes / (1024**2)

print("=== Resultado Questão 10 ===")
print(f"Dados: RDM = {RDM_bits} bits | REM = {REM_bits} bits | Célula = {celula_bits} bits")
print(f"\na) Tamanho do barramento de endereços (BE): {BE_bits} bits")
print(f"   → Porque BE = REM. Eles são ligados diretamente.")
print(f"\nb) Células lidas por operação: {celulas_por_leitura} células")
print(f"   → Porque a palavra/RDM = {RDM_bits} bits e cada célula = {celula_bits} bits")
print(f"   → Cálculo: {RDM_bits} / {celula_bits} = {celulas_por_leitura} células por acesso")
print(f"\nc) Total de bits na memória principal:")
print(f"   → Endereços: 2^{REM_bits} = {N_total_celulas:,} células")
print(f"   → Total: {N_total_celulas:,} × {celula_bits} bits = {T_total_bits:,} bits")
print(f"   → = {T_total_bytes:,.0f} bytes = {T_total_Mbytes} Mbytes = 1 Mbyte")
```
