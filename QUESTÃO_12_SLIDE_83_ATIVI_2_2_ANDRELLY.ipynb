```python
# Questão 12 - Slide 83 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
endereco_hex = "2C81"
conteudo_hex = "F5A"

# Convertendo hexa pra binário pra contar os bits
endereco_bin = bin(int(endereco_hex, 16))
conteudo_bin = bin(int(conteudo_hex, 16))

# Removendo o '0b' do início pra contar só os bits
endereco_bits_minimo = len(endereco_bin) - 2
conteudo_bits = len(conteudo_bin) - 2

# a) Tamanho do REM e RDM
# REM precisa ter bits suficientes pra representar o endereço 2C81
# Como o endereço usa 4 dígitos hexa, o padrão é 4×4 = 16 bits
REM_bits = 16  # 4 dígitos hexa = 16 bits
# RDM = palavra = célula. O conteúdo F5A usa 3 dígitos hexa = 3×4 = 12 bits
RDM_bits = 12  # 3 dígitos hexa = 12 bits

# b) Máxima quantidade de bits na memória
# Se REM = 16 bits, posso endereçar 2^16 células
N_max_celulas = 2 ** REM_bits
T_max_bits = N_max_celulas * RDM_bits
T_max_bytes = T_max_bits / 8
T_max_Kbytes = T_max_bytes / 1024

print("=== Resultado Questão 12 ===")
print(f"Endereço: {endereco_hex} = {endereco_bin} → usa no mínimo {endereco_bits_minimo} bits")
print(f"Conteúdo: {conteudo_hex} = {conteudo_bin} → usa {conteudo_bits} bits")
print(f"\na) Tamanho dos registradores:")
print(f"   → REM: {REM_bits} bits")
print(f"     Porque o endereço {endereco_hex} ocupa 4 dígitos hexa = 4×4 = 16 bits")
print(f"   → RDM: {RDM_bits} bits")
print(f"     Porque o conteúdo {conteudo_hex} ocupa 3 dígitos hexa = 3×4 = 12 bits")
print(f"     E o enunciado diz que célula = palavra = RDM")
print(f"\nb) Máxima quantidade de bits na memória:")
print(f"   → Endereços possíveis: 2^{REM_bits} = {N_max_celulas:,} células")
print(f"   → Total: {N_max_celulas:,} × {RDM_bits} bits = {T_max_bits:,} bits")
print(f"   → = {T_max_bytes:,.0f} bytes = {T_max_Kbytes} Kbytes = 96 Kbytes")
```
