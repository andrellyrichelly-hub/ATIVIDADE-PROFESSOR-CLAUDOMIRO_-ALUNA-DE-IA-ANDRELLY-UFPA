```python
# Questão 5 - Slide 79 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

import math

def analisar_memoria(nome, N_celulas_K, M_bits_por_celula):
    # Converte K pra valor real: 1K = 1024
    N_celulas = N_celulas_K * 1024

    # 1. Endereço: quantos bits preciso pra endereçar N células?
    REM_bits = int(math.log2(N_celulas))
    maior_endereco = N_celulas - 1

    # 2. Conteúdo: o que cabe em cada célula?
    conteudo_bytes = M_bits_por_celula / 8

    # 3. Total de bits
    T_total_bits = N_celulas * M_bits_por_celula
    T_total_bytes = T_total_bits / 8
    T_total_Kbytes = T_total_bytes / 1024

    print(f"=== {nome} ===")
    print(f"Organização: {N_celulas_K}K células de {M_bits_por_celula} bits cada")
    print(f"\n1. ENDEREÇO:")
    print(f"   → REM/BE precisa de {REM_bits} bits")
    print(f"   → Porque 2^{REM_bits} = {N_celulas:,} endereços")
    print(f"   → Maior endereço: {maior_endereco:,} = {bin(maior_endereco)} = {hex(maior_endereco)}")
    print(f"\n2. CONTEÚDO:")
    print(f"   → Cada célula armazena {M_bits_por_celula} bits = {conteudo_bytes} byte(s)")
    print(f"\n3. TOTAL DE BITS:")
    print(f"   → {N_celulas:,} × {M_bits_por_celula} = {T_total_bits:,} bits")
    print(f"   → = {T_total_bytes:,.0f} bytes = {T_total_Kbytes} Kbytes")
    print("-" * 60)

    return T_total_bits, REM_bits

# Analisando as 3 memórias
total_A, REM_A = analisar_memoria("Memória A", 32, 8)
total_B, REM_B = analisar_memoria("Memória B", 16, 16)
total_C, REM_C = analisar_memoria("Memória C", 16, 8)

print("\n=== COMPARAÇÃO FINAL ===")
print(f"Memória A: {REM_A} bits de endereço | 8 bits/célula | {total_A:,} bits totais = 256 Kbits")
print(f"Memória B: {REM_B} bits de endereço | 16 bits/célula | {total_B:,} bits totais = 256 Kbits")
print(f"Memória C: {REM_C} bits de endereço | 8 bits/célula | {total_C:,} bits totais = 128 Kbits")
print("\nCONCLUSÃO:")
print("A e B têm o mesmo total de bits, mas A tem mais endereços e células menores.")
print("A e C têm mesmo tamanho de célula, mas A tem o dobro de endereços e total.")
print("B e C têm mesmo número de endereços, mas B tem células com dobro do tamanho.")
```
