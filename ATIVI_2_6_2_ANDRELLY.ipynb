```python
# Atividade 2.6 - Slide 49: Exemplos 5.5, 5.6 e 5.7
# Cache Associativa por Conjunto - 4-way, 8-way e 32-way
# Andrelly Richelly - Arquitetura com Claudomiro

# Parâmetros base - slide 54 e 55
TAM_MP = 2**32          # 4 GBytes
TAM_BLOCO = 2**6        # 64 Bytes = 6 bits offset
NUM_BLOCOS_MP = 2**26   # 64M Blocos
TAM_CACHE = 2**16       # 64 KBytes
NUM_LINHAS_CACHE = 2**10 # 1024 linhas totais

def mapear_cache_conjunto(endereco_32bits, num_linhas_por_conjunto):
    """
    Mapeamento Associativo por Conjunto
    num_linhas_por_conjunto = 4, 8 ou 32
    """
    offset = endereco_32bits & 0x3F # 6 bits - bits 5-0
    num_conjuntos = NUM_LINHAS_CACHE // num_linhas_por_conjunto
    bits_conjunto = int.bit_length(num_conjuntos - 1)

    conjunto = (endereco_32bits >> 6) & (num_conjuntos - 1)  # bits do conjunto
    tag = endereco_32bits >> (6 + bits_conjunto)            # bits restantes

    nome = f"{num_linhas_por_conjunto}-WAY"
    print(f"5.{4+int.bit_length(num_linhas_por_conjunto)-2} {nome:<8} -> Tag: {tag:06d} | Conjunto: {conjunto:02d} | Offset: {offset:02d} | Bits Conjunto: {bits_conjunto}")
    return tag, conjunto, offset

# --- TESTE AUTOMÁTICO - EXEMPLOS 5.5, 5.6 e 5.7 ---
print("=== ATIVIDADE 2.6 - SLIDE 49: EXEMPLOS 5.5, 5.6 e 5.7 ===\n")
print("MP=4GB, Bloco=64B, Cache=64KB, 1024 Linhas totais\n")

# Endereços de teste
enderecos_teste = [0x00000000, 0x00001040, 0x0FFFFFFF, 0x12345678]

for end in enderecos_teste:
    print(f"Endereço: 0x{end:08X} | {end:032b}")
    # 5.5 = 4-way set associative: 1024/4 = 256 conjuntos = 8 bits
    mapear_cache_conjunto(end, 4)
    # 5.6 = 8-way set associative: 1024/8 = 128 conjuntos = 7 bits
    mapear_cache_conjunto(end, 8)
    # 5.7 = 32-way set associative: 1024/32 = 32 conjuntos = 5 bits
    mapear_cache_conjunto(end, 32)
    print("-" * 80)

print("\nRESUMO DOS 3 EXEMPLOS:")
print("5.5 - 4-WAY:  256 conjuntos | 8 bits conj | 18 bits tag")
print("5.6 - 8-WAY:  128 conjuntos | 7 bits conj | 19 bits tag")
print("5.7 - 32-WAY:  32 conjuntos | 5 bits conj | 21 bits tag")
print("\nTeste concluído com sucesso.")
```
