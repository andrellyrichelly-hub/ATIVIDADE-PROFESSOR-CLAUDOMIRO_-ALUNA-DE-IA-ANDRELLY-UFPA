```python
# Questão 32, 33 E 34 - Slide 87 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# ===== QUESTÃO 32 - CÁLCULO =====
# Dados
linhas_endereco = 16  # 16 linhas = 16 bits de endereço = BE
linhas_saida = 4      # 4 linhas de dados = 4 bits por célula = palavra

# Cálculo
N_celulas = 2 ** linhas_endereco
bits_por_celula = linhas_saida
T_total_bits = N_celulas * bits_por_celula
T_total_bytes = T_total_bits / 8
T_total_Kbytes = T_total_bytes / 1024

print("=== Resultado Questão 32 ===")
print(f"16 linhas de endereçamento → 2^16 = {N_celulas:,} endereços/células")
print(f"4 linhas de saída de dados → {bits_por_celula} bits por célula")
print(f"\nTotal de bits: {N_celulas:,} × {bits_por_celula} = {T_total_bits:,} bits")
print(f"Total de bytes: {T_total_bits:,} ÷ 8 = {T_total_bytes:,.0f} bytes")
print(f"Resposta: {T_total_bytes:,.0f} bytes = {T_total_Kbytes} Kbytes = 32 Kbytes")
print("-" * 60)
```

Questão 32: Resposta

Uma memória com 16 linhas de endereçamento consegue endereçar 2^16 = 65.536 células.
Com 4 linhas de saída de dados, cada célula armazena 4 bits = 0,5 byte.
Total: 65.536 × 4 bits = 262.144 bits = 32.768 bytes = 32 Kbytes.

Sim, considero. A afirmação é válida.
Aumentar a memória RAM reduz a necessidade de acesso ao disco rígido, o que torna o sistema mais rápido.

Quando a RAM está cheia, o sistema usa "memória virtual" no HD, que é 1000x mais lento que RAM. Esse processo se chama swap ou paginação.
Com mais RAM, mais programas e dados ficam carregados direto na memória principal, evitando leituras/escritas no disco.
Então aumentar a RAM não deixa o HD mais rápido, mas diminui a quantidade de vezes que preciso acessar o HD, melhorando o desempenho geral.
É por isso que colocar mais RAM é o upgrade que mais dá ganho de velocidade em PCs antigos.
