```python
# Questão 11 - Slide 82 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

import math

# Dados do problema
N_celulas_K = 32  # 32K células
palavra_bits = 8  # Cada célula = 1 palavra de 8 bits

# Convertendo K pra valor real
N_celulas = N_celulas_K * 1024

# a) Maior endereço em decimal
# Endereços começam em 0, então o maior é N-1
maior_endereco_decimal = N_celulas - 1
maior_endereco_bin = bin(maior_endereco_decimal)
maior_endereco_hex = hex(maior_endereco_decimal)

# b) Tamanho do barramento de endereços (BE)
# BE = REM. Preciso de bits suficientes pra endereçar N_celulas
# Fórmula: N = 2^E → E = log2(N)
BE_bits = REM_bits = int(math.log2(N_celulas))

# c) Bits no RDM e no REM
# RDM tem o tamanho da palavra/célula
RDM_bits = palavra_bits
# REM já calculamos acima

# d) Total máximo de bits na memória
T_total_bits = N_celulas * palavra_bits
T_total_bytes = T_total_bits / 8
T_total_Kbytes = T_total_bytes / 1024

print("=== Resultado Questão 11 ===")
print(f"Dados: {N_celulas_K}K células | {palavra_bits} bits por célula")
print(f"\na) Maior endereço em decimal: {maior_endereco_decimal:,}")
print(f"   → Em binário: {maior_endereco_bin}")
print(f"   → Em hexa: {maior_endereco_hex}")
print(f"   → Porque endereços vão de 0 até {N_celulas:,} - 1")
print(f"\nb) Tamanho do barramento de endereços (BE): {BE_bits} bits")
print(f"   → Porque 2^{BE_bits} = {2**BE_bits:,} endereços = {N_celulas_K}K")
print(f"\nc) Bits armazenados nos registradores:")
print(f"   → RDM: {RDM_bits} bits, porque a palavra tem {palavra_bits} bits")
print(f"   → REM: {REM_bits} bits, porque precisa endereçar {N_celulas:,} células")
print(f"\nd) Total máximo de bits na memória:")
print(f"   → {N_celulas:,} células × {palavra_bits} bits = {T_total_bits:,} bits")
print(f"   → = {T_total_bytes:,.0f} bytes = {T_total_Kbytes} Kbytes = 256 Kbits")
```
