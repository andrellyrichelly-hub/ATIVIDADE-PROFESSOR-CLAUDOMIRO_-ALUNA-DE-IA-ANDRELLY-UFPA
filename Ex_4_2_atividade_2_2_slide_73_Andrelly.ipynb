```python
# Exemplo 4.2 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
T_total_bits = 256 * 1024  # 256K bits = 256 * 1024 = 262.144 bits
M = 8                     # cada célula armazena 8 bits = 1 byte

# 1. Total de células que podem ser utilizadas
N_total_celulas = T_total_bits // M
N_total_Kcelulas = N_total_celulas / 1024

# 2. Tamanho de cada endereço em bits
# Fórmula: N = 2^E, onde E = quantidade de bits do endereço
import math
E_bits_endereco = int(math.log2(N_total_celulas))

print("=== Resultado Exemplo 4.2 ===")
print(f"1. Total de células: {N_total_celulas} células")
print(f"   = {N_total_Kcelulas} K células = 32K células")
print(f"   = {N_total_celulas} bytes, já que 1 célula = 8 bits = 1 byte")
print(f"\n2. Tamanho de cada endereço: {E_bits_endereco} bits")
print(f"   Porque 2^{E_bits_endereco} = {2**E_bits_endereco} endereços")
```
