```python
# Atividade 2.5 - Chip de Memória 32x16 com 4 Bancos
# Andrelly Richelly - Arquitetura com Claudomiro

class BancoMemoria:
    """
    Representa um banco de memória com 32 palavras de 16 bits cada.
    """
    def __init__(self, tamanho=32, largura=16):
        self.tamanho = tamanho
        self.largura = largura
        self.mem = [0] * tamanho

    def ler(self, endereco):
        if 0 <= endereco < self.tamanho:
            return self.mem[endereco]
        else:
            raise ValueError(f"Endereço {endereco} fora do limite 0-{self.tamanho-1}")

    def escrever(self, endereco, dado):
        if 0 <= endereco < self.tamanho:
            mascara = (1 << self.largura) - 1
            self.mem[endereco] = dado & mascara
        else:
            raise ValueError(f"Endereço {endereco} fora do limite 0-{self.tamanho-1}")

class ChipMemoria32x16_4Bancos:
    """
    Chip de memória 32x16 organizado em 4 bancos.
    Total: 128 palavras = 256 bytes
    """
    def __init__(self):
        self.num_bancos = 4
        self.tam_banco = 32
        self.largura = 16
        self.bancos = [BancoMemoria(self.tam_banco, self.largura) for _ in range(self.num_bancos)]

    def decodificar_endereco(self, endereco_global):
        if not 0 <= endereco_global < (self.num_bancos * self.tam_banco):
            raise ValueError(f"Endereço global {endereco_global} fora do limite 0-127")
        banco = (endereco_global >> 5) & 0b11
        endereco_local = endereco_global & 0b11111
        return banco, endereco_local

    def ler(self, endereco_global):
        banco, end_local = self.decodificar_endereco(endereco_global)
        dado = self.bancos[banco].ler(end_local)
        print(f"LEITURA: End_Global={endereco_global:03d} | Banco={banco} | End_Local={end_local:02d} | Dado=0x{dado:04X}")
        return dado

    def escrever(self, endereco_global, dado):
        banco, end_local = self.decodificar_endereco(endereco_global)
        self.bancos[banco].escrever(end_local, dado)
        print(f"ESCRITA: End_Global={endereco_global:03d} | Banco={banco} | End_Local={end_local:02d} | Dado=0x{dado:04X}")

# --- TESTE AUTOMÁTICO - JÁ RODA SOZINHO ---
print("=== ATIVIDADE 2.5 - CHIP 32x16 COM 4 BANCOS ===\n")

chip = ChipMemoria32x16_4Bancos()

print("Escrevendo valores de teste:")
chip.escrever(0, 0xABCD)
chip.escrever(31, 0x1111)
chip.escrever(32, 0x2222)
chip.escrever(64, 0x3333)
chip.escrever(96, 0x4444)
chip.escrever(127, 0xFFFF)

print("\nLendo valores de volta:")
chip.ler(0)
chip.ler(31)
chip.ler(32)
chip.ler(64)
chip.ler(96)
chip.ler(127)

print("\nTeste concluído com sucesso.")
```
