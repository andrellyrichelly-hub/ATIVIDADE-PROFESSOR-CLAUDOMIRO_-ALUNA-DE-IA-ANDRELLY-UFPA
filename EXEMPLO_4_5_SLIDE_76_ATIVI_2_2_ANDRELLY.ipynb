```python
# Exemplo 4.5 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
BE_bits = 33  # Barramento de Endereços = 33 bits → REM = 33 bits
palavras_por_acesso = 4  # BD transfere 4 palavras por acesso
fracao_palavra_por_celula = 1/8  # Cada célula = 1/8 da palavra
T_total_bits = 64 * (1024**3)  # 64G bits = 64 * 2^30 bits

# b) Tamanho do REM
# BE = REM, então REM tem o mesmo tamanho do barramento de endereços
REM_bits = BE_bits

# c) Tamanho da célula e da palavra
# Se REM = 33 bits, então temos 2^33 endereços
N_enderecos = 2 ** REM_bits
# Cada endereço aponta pra 1 célula. Então temos 2^33 células
# Total de bits = N_celulas * bits_por_celula
celula_bits = T_total_bits / N_enderecos
# Se 1 célula = 1/8 da palavra, então palavra = 8 * célula
palavra_bits = celula_bits / fracao_palavra_por_celula

# b) Tamanho do BD
# BD transfere 4 palavras por acesso
BD_bits = palavras_por_acesso * palavra_bits

# a) Quantidade máxima de células
N_total_celulas = N_enderecos  # 1 endereço = 1 célula

print("=== Resultado Exemplo 4.5 ===")
print(f"c) Tamanho de cada célula: {int(celula_bits)} bits")
print(f"   Tamanho da palavra: {int(palavra_bits)} bits")
print(f"   → Porque 1 célula = 1/8 da palavra. Se célula = 8 bits, palavra = 8×8 = 64 bits")
print(f"\nb) Tamanho do REM: {REM_bits} bits")
print(f"   → Igual ao BE, pois REM se conecta direto no barramento de endereços")
print(f"   Tamanho do BD: {int(BD_bits)} bits")
print(f"   → Porque BD transfere 4 palavras: 4 × {int(palavra_bits)} = {int(BD_bits)} bits")
print(f"\na) Quantidade máxima de células: {N_enderecos:,} células")
print(f"   = {N_enderecos / (1024**3)} G células = 8G células")
print(f"   → Cálculo: 2^{REM_bits} = {N_enderecos:,}")
print(f"\nConferindo: {N_enderecos:,} células × {int(celula_bits)} bits = {T_total_bits:,} bits = 64G bits ✓")
```
