```python
# Atividade 2.6 - Exemplos 5.2, 5.3 e 5.4 - Mapeamento de Cache
# Andrelly Richelly - Arquitetura com Claudomiro

# Parâmetros do sistema - slide 37 e 38
TAM_MP = 2**32          # 4 GBytes
TAM_BLOCO = 2**6        # 64 Bytes = 6 bits offset
NUM_BLOCOS_MP = 2**26   # 64M Blocos
TAM_CACHE = 2**16       # 64 KBytes
NUM_LINHAS_CACHE = 2**10 # 1024 linhas

def decodificar_endereco(endereco_32bits, tipo_mapeamento):
    """
    Divide o endereço de 32 bits conforme o tipo de mapeamento
    """
    offset = endereco_32bits & 0x3F # 6 bits - bits 5-0
    bloco_mp = endereco_32bits >> 6 # 26 bits - bits 31-6

    print(f"\nEndereço: {endereco_32bits} | 0x{endereco_32bits:08X} | {endereco_32bits:032b}")
    print(f"Bloco MP: {bloco_mp} | Offset: {offset}")

    if tipo_mapeamento == "DIRETO": # Exemplo 5.2
        # 16 bits tag | 10 bits linha | 6 bits offset
        linha = (endereco_32bits >> 6) & 0x3FF  # bits 15-6
        tag = endereco_32bits >> 16             # bits 31-16
        print(f"5.2 DIRETO      -> Tag: {tag} | Linha: {linha} | Offset: {offset}")
        return tag, linha, None, offset

    elif tipo_mapeamento == "TOTAL_ASSOCIATIVO": # Exemplo 5.3
        # 26 bits tag | 6 bits offset - qualquer linha
        tag = bloco_mp
        print(f"5.3 TOTAL ASSOC -> Tag: {tag} | Linha: Qualquer | Offset: {offset}")
        return tag, None, None, offset

    elif tipo_mapeamento == "CONJ_ASSOCIATIVO": # Exemplo 5.4
        # 21 bits tag | 5 bits conjunto | 6 bits offset
        # 32 conjuntos, 32 linhas por conjunto
        conjunto = (endereco_32bits >> 6) & 0x1F  # bits 10-6
        tag = endereco_32bits >> 11               # bits 31-11
        print(f"5.4 CONJ ASSOC  -> Tag: {tag} | Conjunto: {conjunto} | Offset: {offset}")
        return tag, None, conjunto, offset

# --- TESTE AUTOMÁTICO - EXEMPLOS 5.2, 5.3 e 5.4 ---
print("=== ATIVIDADE 2.6 - EXEMPLOS 5.2, 5.3 e 5.4 ===\n")
print("MP=4GB, Bloco=64B, Cache=64KB, 1024 Linhas")

# Endereços de teste do slide
enderecos_teste = [0x00000000, 0x04000040, 0x0FFFFFFF, 0x12345678]

for end in enderecos_teste:
    decodificar_endereco(end, "DIRETO")
    decodificar_endereco(end, "TOTAL_ASSOCIATIVO")
    decodificar_endereco(end, "CONJ_ASSOCIATIVO")
    print("-" * 70)

print("\nTeste concluído com sucesso.")
```
