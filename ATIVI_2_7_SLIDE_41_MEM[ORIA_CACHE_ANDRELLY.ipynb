```python
# Atividade 2.7 Aluna: Andrelly Richelly Disciplina: Arquitetura de computadores
# Simulador de Memória Cache - Mapeamento Direto | Extensão da Atividade 2.1
# Prof. Claudomiro | Baseado no Slide 41

# ===================================================================
# 1. CONFIGURAÇÃO E ESTRUTURAS DE DADOS
# ===================================================================
TAMANHO_RAM = 16 # Endereços 0000 a 1111, igual à 2.1
TAMANHO_CACHE = 4 # 4 linhas, conforme exemplo do slide 41
BITS_POR_LINHA = 8 # 8 bits de dado por posição

# Memória RAM: dicionário {endereco: dado}
memoria_ram = {format(i, '04b'): "00000000" for i in range(TAMANHO_RAM)}

# Memória Cache: lista de listas [tag, dado, bit_validade]
# bit_validade = 0 significa linha vazia/inválida
cache = [["", "00000000", 0] for _ in range(TAMANHO_CACHE)]

# Contadores globais para estatísticas
hits = 0
misses = 0

# ===================================================================
# 2. FUNÇÕES DO SIMULADOR
# ===================================================================
def mostrar_ram():
    print("--- MEMÓRIA RAM PRINCIPAL ---")
    for endereco, dado in sorted(memoria_ram.items()):
        print(f"Endereço {endereco}: {dado}")
    print("-" * 30)

def mostrar_cache():
    print("\n--- MEMÓRIA CACHE | Mapeamento Direto ---")
    print("Linha | Válido | Tag | Dado")
    for i, linha in enumerate(cache):
        tag, dado, valido = linha[0], linha[1], linha[2]
        print(f" {i} | {valido} | {tag:>4} | {dado}")
    print("-" * 35)

def gravar_na_ram(endereco, dado):
    """Grava um dado de 8 bits em um endereço da RAM."""
    if endereco in memoria_ram and len(dado) == BITS_POR_LINHA:
        memoria_ram[endereco] = dado
        print(f"[RAM] Dado {dado} gravado no endereço {endereco}")
    else:
        print(f"ERRO: Endereço {endereco} inválido ou dado não tem 8 bits.")

def ler_dado(endereco):
    """Lê um dado. Primeiro busca na cache. Se não achar, busca na RAM."""
    global hits, misses

    if endereco not in memoria_ram:
        print(f"ERRO: Endereço {endereco} não existe na RAM.")
        return None

    # Mapeamento Direto: indice = endereco % tamanho_cache
    indice_cache = int(endereco, 2) % TAMANHO_CACHE
    # Tag: bits do endereço que não são usados no índice. Pra 4 linhas, usa 2 bits de índice.
    tag = endereco[:-2]

    linha_cache = cache[indice_cache]
    tag_na_cache, dado_na_cache, bit_valido = linha_cache[0], linha_cache[1], linha_cache[2]

    # Verifica CACHE HIT: linha válida E tags iguais
    if bit_valido == 1 and tag_na_cache == tag:
        hits += 1
        print(f"[CACHE HIT] Endereço {endereco} encontrado na Linha {indice_cache} da cache.")
        return dado_na_cache

    # CACHE MISS: busca na RAM e atualiza a cache
    else:
        misses += 1
        print(f"[CACHE MISS] Endereço {endereco} não está na cache. Acessando RAM...")
        dado_da_ram = memoria_ram[endereco]

        # Política de substituição: sobrescreve a linha
        cache[indice_cache] = [tag, dado_da_ram, 1]
        print(f"[CACHE UPDATE] Linha {indice_cache} atualizada: Tag={tag}, Dado={dado_da_ram}")
        return dado_da_ram

def mostrar_estatisticas():
    """Mostra a taxa de hit e miss da cache."""
    total_acessos = hits + misses
    if total_acessos == 0:
        print("\nNenhum acesso à memória foi realizado ainda.")
        return

    taxa_hit = (hits / total_acessos) * 100
    print("\n--- ESTATÍSTICAS DE DESEMPENHO DA CACHE ---")
    print(f"Total de leituras: {total_acessos}")
    print(f"Cache Hits: {hits}")
    print(f"Cache Misses: {misses}")
    print(f"Taxa de Hit: {taxa_hit:.2f}%")
    print("-" * 42)

# ===================================================================
# 3. EXECUÇÃO E DEMONSTRAÇÃO - IGUAL AO SLIDE 41
# ===================================================================
print("="*55)
print("SIMULADOR DE CACHE - ATIVIDADE 2.7 - ANDRELLY RICHELLY")
print("="*55)

# Estado inicial
mostrar_ram()
mostrar_cache()

# Gravando dados na RAM para teste
print("\nGravando dados iniciais na RAM...")
gravar_na_ram("0000", "10101010") # End 0 -> Linha 0
gravar_na_ram("0001", "11110000") # End 1 -> Linha 1
gravar_na_ram("0100", "00110011") # End 4 -> Linha 0, conflito com 0000
gravar_na_ram("1000", "11001100") # End 8 -> Linha 0, conflito com 0000 e 0100

mostrar_ram()

# Sequência de leituras para demonstrar Hit, Miss e Conflito
print("\n--- INICIANDO SEQUÊNCIA DE LEITURAS ---")

print("\n1. Lendo endereço 0000 pela primeira vez:")
ler_dado("0000") # MISS

print("\n2. Lendo endereço 0000 novamente:")
ler_dado("0000") # HIT

print("\n3. Lendo endereço 0100:")
ler_dado("0100") # MISS, sobrescreve a linha 0

print("\n4. Lendo endereço 0000 de novo:")
ler_dado("0000") # MISS, pois foi sobrescrito. Demonstra a desvantagem!

print("\n5. Lendo endereço 1000:")
ler_dado("1000") # MISS, sobrescreve a linha 0 novamente

# Resultado final
mostrar_cache()
mostrar_estatisticas()
```

### **Explicação dos Resultados - Slide 41**

1. **Cache Miss Inicial**: Toda primeira leitura de um endereço gera miss porque a cache começa vazia com bit de validade = 0.

2. **Cache Hit**: A segunda leitura do endereço 0000 deu hit. A CPU economizou tempo por não acessar a RAM.

3. **Conflito do Mapeamento Direto**: Os endereços 0000, 0100 e 1000 todos caem na Linha 0 da cache porque 0 % 4 = 4 % 4 = 8 % 4 = 0.

   Quando lemos 0100, ele expulsa o dado de 0000 da cache. Por isso a quarta leitura de 0000 vira miss de novo. Essa é a principal desvantagem do mapeamento direto.

4. **Taxa de Hit**: A taxa ficou baixa neste exemplo porque forçamos os conflitos. Em programas reais, a Taxa de Hit passa de 90% devido ao Princípio da Localidade Temporal e Espacial.
