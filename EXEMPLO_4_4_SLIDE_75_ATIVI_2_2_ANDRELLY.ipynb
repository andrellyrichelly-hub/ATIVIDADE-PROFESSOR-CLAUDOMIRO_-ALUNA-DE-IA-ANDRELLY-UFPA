```python
# Exemplo 4.4 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

# Dados do problema
REM_bits = 24  # Registrador de Endereços: 24 bits
RDM_bits = 32  # Registrador de Dados: 32 bits
celulas_por_acesso = 2  # Em cada acesso, lê 2 células

# c) Tamanho da palavra e de cada célula
# RDM = tamanho da palavra = tamanho do barramento de dados
palavra_bits = RDM_bits
# Se a palavra tem 32 bits e vem de 2 células, cada célula tem:
celula_bits = palavra_bits // celulas_por_acesso

# a) Capacidade máxima de endereçamento
# REM define quantos endereços existem: N = 2^REM_bits
N_total_enderecos = 2 ** REM_bits
N_total_M_enderecos = N_total_enderecos / (1024**2)  # Converte pra Mega

# b) Total máximo de bits na RAM
# Cada endereço aponta pra 1 palavra, mas a palavra vem de 2 células físicas
# Mas pro cálculo de capacidade total, usamos: N_enderecos * tamanho_da_celula
T_total_bits = N_total_enderecos * celula_bits
T_total_Gbits = T_total_bits / (1024**3)
T_total_Gbytes = T_total_bits / 8 / (1024**3)

print("=== Resultado Exemplo 4.4 ===")
print(f"c) Tamanho da palavra: {palavra_bits} bits")
print(f"   Tamanho de cada célula: {celula_bits} bits")
print(f"   → Porque RDM = 32 bits = palavra, e cada acesso lê 2 células: 32/2 = 16 bits")
print(f"\na) Capacidade máxima de endereçamento: {N_total_enderecos:,} endereços")
print(f"   = {N_total_M_enderecos} M endereços = 16M endereços")
print(f"   → Porque REM tem {REM_bits} bits: 2^{REM_bits} = {2**REM_bits:,}")
print(f"\nb) Total máximo de bits na RAM: {T_total_bits:,} bits")
print(f"   = {T_total_Gbits} Gbits = {T_total_Gbytes} Gbytes = 32 Mbytes")
print(f"   → Cálculo: {N_total_enderecos:,} endereços × {celula_bits} bits/célula")
```
