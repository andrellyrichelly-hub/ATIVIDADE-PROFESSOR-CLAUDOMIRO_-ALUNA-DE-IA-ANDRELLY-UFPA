```python
# Questões 27 e 28 - Slide 85 - Atividade 2.2 - Arquitetura de Computadores
# Aluna: Andrelly Richelly

import math

# Função pra calcular bits de endereço - uso nas duas questões
def calcular_bits_endereco(tamanho, unidade):
    if unidade == "G":
        N_celulas = tamanho * (1024**3)  # 1G = 2^30
    elif unidade == "K":
        N_celulas = tamanho * 1024       # 1K = 2^10
    elif unidade == "M":
        N_celulas = tamanho * (1024**2)  # 1M = 2^20
    else:
        N_celulas = tamanho

    # Fórmula: N = 2^E → E = log2(N)
    E_bits = int(math.log2(N_celulas))

    return N_celulas, E_bits

# Questão 27: Memória de 128G células
N_128G, E_128G = calcular_bits_endereco(128, "G")

# Questão 28: Memória de 32K células
N_32K, E_32K = calcular_bits_endereco(32, "K")

print("=== Resultado Questão 27 ===")
print(f"Memória: 128G células")
print(f"Cálculo: 128 × 2^30 = {N_128G:,} células")
print(f"Bits de endereço necessários: log₂({N_128G:,}) = {E_128G} bits")
print(f"Conferindo: 2^{E_128G} = {2**E_128G:,}")
print(f"Resposta: São necessários {E_128G} bits para endereçar 128G células\n")

print("=== Resultado Questão 28 ===")
print(f"Memória: 32K células")
print(f"Cálculo: 32 × 2^10 = {N_32K:,} células")
print(f"Bits de endereço necessários: log₂({N_32K:,}) = {E_32K} bits")
print(f"Conferindo: 2^{E_32K} = {2**E_32K:,}")
print(f"Resposta: São necessários {E_32K} bits para endereçar 32K células")

print("\n=== Tabela de Referência ===")
print("1K = 2^10 = 1.024 → 10 bits")
print("1M = 2^20 = 1.048.576 → 20 bits")
print("1G = 2^30 = 1.073.741.824 → 30 bits")
print("128G = 2^7 × 2^30 = 2^37 → 37 bits")
print("32K = 2^5 × 2^10 = 2^15 → 15 bits")
```
