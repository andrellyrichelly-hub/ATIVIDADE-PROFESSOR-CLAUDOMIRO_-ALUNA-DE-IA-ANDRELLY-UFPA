Questão 6: Função do REM e RDM


REM - Registrador de Endereços da Memória

- Função: Armazenar o endereço da posição de memória a ser acessada.

- Tamanho: Igual ao barramento de endereços.

- Direção: Unidirecional, do processador para a memória.



RDM - Registrador de Dados da Memória

- Função: Armazenar temporariamente o dado que foi lido da memória ou que será escrito nela.

- Tamanho: Igual ao tamanho da palavra/célula da memória.

- Direção: Bidirecional, entre processador e memória.


Questão 7: Barramentos Processador ↔ MP

Barramento de Endereços (BE)

- Função: Conduz o endereço da MP.
- Direção: Unidirecional (CPU → MP).
- Define: Capacidade máxima de endereçamento.

Barramento de Dados (BD)

- Função: Conduz dados e instruções.
- Direção: Bidirecional (CPU ↔ MP).
- Define: Tamanho da palavra, desempenho de transferência.

Barramento de Controle (BC)

- Função: Conduz sinais de controle como Read, Write, Clock.
- Direção: Bidirecional (CPU ↔ MP e demais dispositivos).
- Define: Tipos de operações possíveis.
